## Centrality Analysis

```python
nx.degree_centrality(fs_graph)
```

```python
most_influential = nx.degree_centrality(fs_graph)
```

```python
for w in sorted(most_influential, key=most_influential.get, reverse=True):
    print(w, most_influential[w])
```

```python
### Most Important Connection
nx.eigenvector_centrality(fs_graph)
```

```python
most_important_link = nx.eigenvector_centrality(fs_graph)
```

```python
for w in sorted(most_important_link, key=most_important_link.get, reverse=True):
    print(w, most_important_link[w])
```

```python
nx.betweenness_centrality(fs_graph)
```

```python
def draw(G, pos, measures, measure_name):
    
    nodes = nx.draw_networkx_nodes(G, pos, node_size=250, cmap=plt.cm.plasma, 
                                   node_color=list(measures.values()),
                                   nodelist=measures.keys())
    nodes.set_norm(mcolors.SymLogNorm(linthresh=0.01, linscale=1, base=10))
    # labels = nx.draw_networkx_labels(G, pos)
    edges = nx.draw_networkx_edges(G, pos)

    plt.title(measure_name)
    plt.colorbar(nodes)
    plt.axis('off')
    plt.show()
```

```python
pos = nx.spring_layout(fs_graph, seed=675)
```

```python
plt.figure(figsize=(20,20))
draw(fs_graph, pos, nx.harmonic_centrality(fs_graph), 'Harmonic Centrality')
```

```python
plt.figure(figsize=(20,20))
draw(fs_graph, pos, nx.degree_centrality(fs_graph), 'Degree Centrality')
```

```python
plt.figure(figsize=(20,20))
draw(fs_graph, pos, nx.betweenness_centrality(fs_graph), 'Betweenness Centrality')
```

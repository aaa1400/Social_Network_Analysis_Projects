```python
# Load Pkgs
import pandas as pd
import networkx as nx
import warnings
```

```python
df = pd.read_excel("TrainTehranNetwork.xlsx")
df
```

```python
df=df.drop('Unnamed: 3',axis=1)
df
```

```python
# Read Our Edgelist
Tr_graph = nx.from_pandas_edgelist(df,source="From",target="To", edge_attr='Line')
pos = nx.spring_layout(Tr_graph)
```

```python
most_influential = nx.degree_centrality(Tr_graph)
```

```python
for w in sorted(most_influential, key=most_influential.get, reverse=True):
    print(w, most_influential[w])
```

```python
### Association 
nx.clustering(Tr_graph)
```

```python
nx.clique.graph_clique_number(Tr_graph, cliques=None)
```

```python
nx.node_connected_component(Tr_graph,"MeydaneShohada")
```

```python
bet_centrality = nx.betweenness_centrality(Tr_graph, normalized = True, 
                                              endpoints = False)
  
# G is the Karate Social Graph, parameters normalized
# and endpoints ensure whether we normalize the value
# and consider the endpoints respectively.
print(bet_centrality)
```

```python
# extract the edge weight
edge_colors = [a['Line'] for u,v,a in Tr_graph.edges(data=True)]
```

```python
# draw nodes and edges separately. allows using colormap for edges.
import numpy as np
plt.figure(figsize=(25,25))
nx.draw_networkx_nodes(Tr_graph, pos=pos)
nx.draw_networkx_edges(Tr_graph, pos=pos, edge_color=edge_colors, edge_cmap=plt.cm.viridis, edge_vmin=1, edge_vmax=np.max(edge_colors), width=5)
nx.draw_networkx_labels(Tr_graph, pos=pos);
plt.show()
```

```python
def draw(G, pos, measures, measure_name):
    
    nodes = nx.draw_networkx_nodes(G, pos, node_size=250, cmap=plt.cm.plasma, 
                                   node_color=list(measures.values()),
                                   nodelist=measures.keys())
    nodes.set_norm(mcolors.SymLogNorm(linthresh=0.01, linscale=1, base=10))
    # labels = nx.draw_networkx_labels(G, pos)
    edges = nx.draw_networkx_edges(G, pos)

    plt.title(measure_name)
    plt.colorbar(nodes)
    plt.axis('off')
    plt.show()
```

```python
pos = nx.spring_layout(an_graph, seed=675)
```

```python
plt.figure(figsize=(20,20))
draw(an_graph, pos, nx.degree_centrality(an_graph), 'Degree Centrality')
```

```python
plt.figure(figsize=(20,20))
draw(an_graph, pos, nx.eigenvector_centrality(an_graph), 'Eigenvector Centrality')
```

```python
plt.figure(figsize=(20,20))
draw(an_graph, pos, nx.closeness_centrality(an_graph), 'Closeness Centrality')
```
